```python
data = data.rename(columns={'Year_of_Birth':'birth_year', 'Precinct_Number':'precinc_id', 'Name':'name', 'Address':'address'})
data.head(1)
```

```python
import datetime
current_year = datetime.datetime.now().year
analysis_result_year = 2016
year=analysis_result_year
```

```python
data['age'] = year - data['birth_year']
data.head()
```

```python
head = data.sort_values('age', ascending=False).head()
head
```

```python
head.reset_index(drop=True, inplace=True)
head
```

```python
head.drop_duplicates(inplace=True)
head
```

```python
oldest_age = head.loc[0,'age']
print('Oldest voter is', oldest_age)
```

```python
youngest_age = data.sort_values(by='age', ascending=True).reset_index(drop=True)['age'][0]
print('Youngest voter is', youngest_age)
```

```python
oldest_voters = data[data['age']==oldest_age]
youngest_voters = data[data['age']==youngest_age]

pd.concat([oldest_voters, youngest_voters])
```

```python
import numpy as np
```

```python
num = 6

bins = np.linspace(youngest_age, oldest_age, num=num)
bins
```

```python
labels = []

for index, value in enumerate(bins):
    value = int(value)
    if index == 0:
        initial_age = value
        continue
    labels.append(f"{initial_age}-{value}")
    initial_age = value

print(bins)
print(labels)
```

```python
data['age_group'] = pd.cut(data['age'], bins=bins, labels=labels, include_lowest=True)
data.head()
```

```python
age_group_table = data.groupby('age_group').size().reset_index(name='count')

# age_group_table['count'] = age_group_table['count'].apply(lambda x: f"{x:,}")

age_group_table
```

```python
total_voters = data.count()['name']
print(f"Total number of voters is {total_voters}")
```

```python
age_group_table['percent_per_total'] = age_group_table['count']/total_voters
age_group_table
```

```python
import matplotlib.pyplot as plt
```

```python
age_group_table.plot.pie(
    y='count',
    labels=age_group_table['age_group'],
    autopct="%.2f%%",
    legend=False
)
plt.ylabel("")
plt.figtext(0.5, -0.05, "Age Groups", ha="center")
plt.show()
```
